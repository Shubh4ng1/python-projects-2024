# Plotting a line chart to view daily cases

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv(r"C:\Users\ACER\Desktop\Datasets for matplotlib Covid-19\case_time_series (2).csv")#reading the file
data.head(10) #viewing the data set summary first 10 values
```

```python
data.shape  #viewing the shape
data.columns #viewing the columns
data.info() #viewing the related information
```

```python
#data above expresses that there are seven columns
#the column names are mentioned from 0-6
#they contain 105 non-null values of data type int64
#entries range from 0-104
```

```python
#assigning values

x=data.iloc[61:,0] #storing the dates
y=data.iloc[61:,1] #total daily confirmed cases
rec= data.iloc[61:,5] #total daily rcovered cases
dec= data.iloc[61:,5] #total daily deceased cases
```

```python
ax= plt.figure(figsize=(25,8)) #width is 25, height s 8
ax=plt.axes()
ax.grid(linewidth=0.4, color='#8f8f8f') #implementing grids
ax.set_xlabel('\nDate',size=25) #setting up x label, and size
ax.set_ylabel('Number of Confirmed Cases\n',size=25) #setting up y label, and size
ax.plot(x,y, color='#1F77B4',marker='o',linewidth=4,markersize=15)

#fixing the overlap by making the labels vertical in x-axis
plt.xticks(rotation='vertical',size='25')
plt.yticks(size='25')

#annotating with loop
for i,j in zip(x,y):
 ax.annotate(str(j), xy=(i,j+500), size='20')

plt.title("COVID-19 IN : Daily Confirmed\n",size=40,color='#28a9ff')#graph heading
```

# Plotting a pie chart to view the distribution of various factors that contributed to the transmission

```python
# considering that factors such as travel, places, contact,and others aided transmission a pie chart was made

part=[25,22,56,41]
factors=['travel', 'places', 'contact', 'others']
color= ['#567fd9','#8cb3ee','#007889','#006666']
exp= [0.2,0.2,0.2,0.2]


#defining pie chart features, parts, label names,radius colors and more
plt.pie(part,labels=factors,
        textprops=dict(size=25,color='black'),
        radius=3,
        colors=color,
        autopct='%2.2f%%',
        explode=exp,
        shadow=True,
        startangle=120) 
plt.title('Transmission distribution',color='black',size=40) #chart title
```

# Viewing district wise cases for a state using a bar plot 

```python
#reading the second dataset
dataa=pd.read_csv(r'C:\Users\ACER\Desktop\Datasets for matplotlib Covid-19\district.csv')
dataa.head() #viewing top 5 records
```

```python
dataa.shape  #viewing the shape
dataa.columns #viewing the columns
dataa.info() #viewing the related information
```

```python
#data above expresses that there are six columns
#the column names are mentioned from 0-5
#they mostly contain 34 non-null values of data type float64
#entries range from 0-37
```

```python
dist=list(data.iloc[:30,0]) #district names
rcov=dataa.iloc[:30,5].values #total recovered counts of patients
dczd=dataa.iloc[:30,4].values #total deceased counts of patients
cfrd=dataa.iloc[:30,3].values #total confirmed counts of patients

plt.figure(figsize=(25,10))
ax=plt.axes()
ax.set_facecolor('yellow')
ax.grid(linewidth=0.4, color='#8f8f8f')

plt.xticks(rotation='vertical',size='20',color='black')#ticks of X

plt.yticks(size='20',color='black') #ticks of y

ax.set_xlabel('\nDistrict',size=25,color='black')
ax.set_ylabel('No. of cases\n',size=25,color='black')
plt.tick_params(size=20,color='white')


ax.set_title('District wise breakdown\n',
             size=50,color='#28a9ff')

plt.bar(dist,cfrd,label='re',color='green')
plt.bar(dist,rcov,label='re',color='blue')
plt.bar(dist,dczd,label='re',color='red')
 
for i,j in zip(dist,rcov):
    ax.annotate(str(int(j)), xy=(i,j+3), color='black',
                size='15')
 
plt.legend(['Confirmed','Recovered','Deceased'],
           fontsize=20)
```
